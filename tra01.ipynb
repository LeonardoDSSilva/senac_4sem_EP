```python
# IMPORTACAO DE BIBLIOTECAS
import os
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
## CARREGAMENTO DE DADOS
df_desmatamento = pd.read_csv('desmatamento_prodes.csv')
## TRATAMENTO DE DADOS - RENOMEAR COLUNA: referencia -> ano 
df_desmatamento.rename(columns={'referencia':'ano'}, inplace=True)

# SOMENTE ANO E AREA DESMATADA
df_desmatamento_ana_total = df_desmatamento[['ano', 'area_total_desmatamento']]
```

```python
# calculo da media, mediana, moda, desvio padrao e variancia - apresentacao em formato de tabela
df_calculos = pd.DataFrame({ 'media': df_desmatamento.mean(),
							 'mediana': df_desmatamento.median(),
							 'moda': df_desmatamento.mode().T[0],
							 'desvio padrao': df_desmatamento.std(),
							 'variancia': df_desmatamento.var()
							})

# remover a linha 'area_total_desmatamento'
df_calculos.drop(index='area_total_desmatamento', inplace=True)
df_calculos.drop(index='ano', inplace=True)
df_calculos.style.format("{:.2f}")
```

```python
# remomver a coluna 'area_total_desmatamento'
df_desmatamento.drop(columns=['area_total_desmatamento'], inplace=True)
```

# GRAFICOS

```python
# percentual de desmatamento de cada estado em relacao ao total
df_desmatamento_percentual = df_desmatamento.copy()
df_desmatamento_percentual.set_index('ano', inplace=True)
df_desmatamento_percentual = df_desmatamento_percentual.div(df_desmatamento_percentual.sum(axis=1), axis=0) * 100
df_desmatamento_percentual.reset_index(inplace=True)

# GRAFICO DE PIZZA
fig = go.Figure()
fig.add_trace(go.Pie(labels=df_desmatamento.columns[1:], values=df_desmatamento.sum()[1:]))
fig.update_layout(title='Desmatamento Prodes por Estado em Percentual')
fig.show()
```

>Pará e Mato Grosso lideram:
- Esses dois estados têm as maiores porcentagens de desmatamento, representando uma parcela significativa do total. Isso é preocupante, pois o desmatamento tem sérias consequências ambientais, como perda de biodiversidade, impacto no clima e degradação do solo.
>Amazonas e Rondônia:
- Também apresentam altos índices: Esses estados também contribuem significativamente para o desmatamento. A região amazônica é especialmente vulnerável devido à sua rica biodiversidade e à importância global na regulação climática.
>Outros estados têm menor impacto:
- Alguns estados, como Acre, Amapá e Tocantins, têm porcentagens menores de desmatamento. No entanto, mesmo esses números mais baixos ainda são motivo de preocupação.

```python
#  percentual de desmatamento em relacao ao ano
df_desmatamento_percentual_ano = df_desmatamento.copy()
df_desmatamento_percentual_ano.set_index('ano', inplace=True)
df_desmatamento_percentual_ano = df_desmatamento_percentual_ano.div(df_desmatamento_percentual_ano.sum(axis=0), axis=1) * 100
df_desmatamento_percentual_ano.reset_index(inplace=True)

# GRAFICO DE PIZZA
fig = go.Figure()
fig.add_trace(go.Pie(labels=df_desmatamento_percentual_ano['ano'], values=df_desmatamento.sum()[1:]))
fig.update_layout(title='Desmatamento Prodes por Ano em Percentual')
fig.show()

```

>Tendência Crescente: O desmatamento parece ter aumentado ao longo dos anos, com um pico notável em 1994 e 1995.

>Anos Críticos: Os anos de 1994 e 1995 representam períodos de alta atividade de desmatamento, indicando possíveis fatores que contribuíram para essa tendência.

```python
## EVOLUCAO DO DESMATAMENTO AO LONGO DOS ANOS
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_desmatamento['ano'], y=df_desmatamento['area_total_desmatamento'], mode='lines+markers'))
fig.update_layout(title='Evolução do Desmatamento na Amazônia Legal (1988-2022)', xaxis_title='Ano', yaxis_title='Área Desmatada (km²)')
fig.show()
```

Este gráfico representa a evolução do desmatamento na Amazônia Legal de 1988 a 2022.

>Tendência Geral:
- Observa-se uma tendência de aumento no desmatamento até o ano de 1995, quando atingiu um pico significativo.
- Após esse período, houve variações, mas a tendência geral foi de redução na área desmatada.
>Picos e Quedas:
- O pico de desmatamento ocorreu em 1995, com uma área significativa desmatada.
- A partir de então, houve reduções em alguns anos, mas também aumentos em outros.

## GRAFICO DE LINHA

```python
## GRAFICO DE LINHAS DE TODAS DE DESMATAMENTO POR ESTADO
def linha(df, x, title):
	fig = go.Figure()
	for coluna in df.columns:
		if coluna != x:
			fig.add_trace(go.Scatter(x=df[x], y=df[coluna], mode='lines', name=coluna))
	fig.update_layout(title=title)
	fig.show()

linha(df_desmatamento, 'ano', 'Desmatamento por estado')
```

#### Este gráfico apresenta as taxas de desmatamento por estado no Brasil de 1988 a 2022. 
- Vemos flutuações significativas e picos, especialmente em Mato Grosso e Pará, destacando a questão ambiental contínua da destruição do habitat. 

## GRAFICO BOX PLOT

```python
# GRAFICO DE BOXPLOT DE DESMATAMENTO POR ESTADO
def box(df, title):
	fig = go.Figure()
	for coluna in df.columns:
		if coluna != 'ano':
			fig.add_trace(go.Box(y=df[coluna], name=coluna))
	fig.update_layout(title=title)
	fig.show()
	
box(df_desmatamento, 'Box Plot do Desmatamento por Estado')
```

>Mato Grosso com Maior Desmatamento:
- O estado de Mato Grosso apresenta níveis significativamente mais altos de desmatamento em comparação com os outros estados.
- A mediana da área desmatada em Mato Grosso é consideravelmente maior do que nos demais estados.
>Variação entre os Estados:
- Os estados de Pará, Amazonas e Rondônia também têm caixas (boxes) mais altas, indicando áreas desmatadas consideráveis.
- Por outro lado, Acre, Amapá, Maranhão, Roraima e Tocantins têm medianas mais baixas, sugerindo menor desmatamento.

## GRAFICO DE BARRAS POR ESTADO

```python
# GRAFICO DE BARRAS
def barras(df, x, y, title):
	fig = px.bar(df, x=x, y=y, title=title)
	fig.show()
	
barras(df_desmatamento, 'ano', 'para', 'DESMATAMENTO PRODES POR ANO NO PARA')
for coluna in df_desmatamento.columns:
	if coluna != 'ano':
		barras(df_desmatamento, 'ano', coluna, f'Desmatamento Prodes por Ano em {coluna.upper()}')
```

## GRAFICO DE PIZZA

```python
# percentual de desmatamento cada parte sera estados por ano
def pizza(df, title):
	fig = px.pie(df, values='desmatamento', names='estado', title=f'Distribuição do desmatamento por estado em {title}')
	fig.show()

for ano in df_desmatamento['ano'].unique():
	df_ano = df_desmatamento[df_desmatamento['ano'] == ano]
	df_melted  = df_ano.melt(id_vars='ano', var_name='estado', value_name='desmatamento')

	pizza(df_melted , ano)
```
