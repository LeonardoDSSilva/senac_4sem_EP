```python
# calculo da media, mediana, moda, desvio padrao e variancia - apresentacao em formato de tabela
df_calculos = pd.DataFrame({ 'media': df_desmatamento.mean(),
							 'mediana': df_desmatamento.median(),
							 'moda': df_desmatamento.mode().T[0],
							 'desvio padrao': df_desmatamento.std(),
							 'variancia': df_desmatamento.var()
							})

# remover a linha 'area_total_desmatamento'
df_calculos.drop(index='area_total_desmatamento', inplace=True)
df_calculos.drop(index='ano', inplace=True)
df_calculos.style.format("{:.2f}")
```

# GRAFICOS

```python
# percentual de desmatamento de cada estado em relacao ao total
df_desmatamento_percentual = df_desmatamento.copy()
df_desmatamento_percentual.set_index('ano', inplace=True)
df_desmatamento_percentual = df_desmatamento_percentual.div(df_desmatamento_percentual.sum(axis=1), axis=0) * 100
df_desmatamento_percentual.reset_index(inplace=True)

# GRAFICO DE PIZZA
fig = go.Figure()
fig.add_trace(go.Pie(labels=df_desmatamento.columns[1:], values=df_desmatamento.sum()[1:]))
fig.update_layout(title='Desmatamento Prodes por Estado em Percentual')
fig.show()
```

## GRAFICO DE LINHA

```python
## GRAFICO DE LINHAS DE TODAS DE DESMATAMENTO POR ESTADO
def linha(df, x, title):
	fig = go.Figure()
	for coluna in df.columns:
		if coluna != x:
			fig.add_trace(go.Scatter(x=df[x], y=df[coluna], mode='lines', name=coluna))
	fig.update_layout(title=title)
	fig.show()

linha(df_desmatamento, 'ano', 'Desmatamento por estado')
```

#### Este grÃ¡fico apresenta as taxas de desmatamento por estado no Brasil de 1988 a 2022. 
- Vemos flutuaÃ§Ãµes significativas e picos, especialmente em Mato Grosso e ParÃ¡, destacando a questÃ£o ambiental contÃ­nua da destruiÃ§Ã£o do habitat. 

## GRAFICO BOX PLOT

```python
# GRAFICO DE BOXPLOT DE DESMATAMENTO POR ESTADO
def box(df, title):
	fig = go.Figure()
	for coluna in df.columns:
		if coluna != 'ano':
			fig.add_trace(go.Box(y=df[coluna], name=coluna))
	fig.update_layout(title=title)
	fig.show()
	
box(df_desmatamento, 'Box Plot do Desmatamento por Estado')
```

>Mato Grosso com Maior Desmatamento:
- O estado de Mato Grosso apresenta nÃ­veis significativamente mais altos de desmatamento em comparaÃ§Ã£o com os outros estados.
- A mediana da Ã¡rea desmatada em Mato Grosso Ã© consideravelmente maior do que nos demais estados.
>VariaÃ§Ã£o entre os Estados:
- Os estados de ParÃ¡, Amazonas e RondÃ´nia tambÃ©m tÃªm caixas (boxes) mais altas, indicando Ã¡reas desmatadas considerÃ¡veis.
- Por outro lado, Acre, AmapÃ¡, MaranhÃ£o, Roraima e Tocantins tÃªm medianas mais baixas, sugerindo menor desmatamento.

## GRAFICO DE BARRAS POR ESTADO

```python
# GRAFICO DE BARRAS
def barras(df, x, y, title):
	fig = px.bar(df, x=x, y=y, title=title)
	fig.show()
	
barras(df_desmatamento, 'ano', 'para', 'DESMATAMENTO PRODES POR ANO NO PARA')
for coluna in df_desmatamento.columns:
	if coluna != 'ano':
		barras(df_desmatamento, 'ano', coluna, f'Desmatamento Prodes por Ano em {coluna.upper()}')
```

## GRAFICO DE PIZZA

```python
# percentual de desmatamento cada parte sera estados por ano
def pizza(df, title):
	fig = px.pie(df, values='desmatamento', names='estado', title=f'DistribuiÃ§Ã£o do desmatamento por estado em {title}')
	fig.show()

for ano in df_desmatamento['ano'].unique():
	df_ano = df_desmatamento[df_desmatamento['ano'] == ano]
	df_melted  = df_ano.melt(id_vars='ano', var_name='estado', value_name='desmatamento')

	pizza(df_melted , ano)
```
